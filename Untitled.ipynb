```python
from ultralytics import YOLO
import cv2

# Load the YOLO model
model = YOLO("best.pt")

# Open the video file
video_path = "downloads/34.mp4"
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create a VideoWriter object
output_path = "output.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Dictionary to store detected objects (bounding boxes) with a timeout
persistent_boxes = {}
timeout_frames = 10  # Keep bounding boxes for 10 frames even if they disappear

frame_count = 0

# Process each frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Stop when video ends

    frame_count += 1

    # Run YOLO inference on the frame
    results = model(frame, conf=0.1)

    # Store new detections
    current_boxes = []
    for result in results:
        for box in result.boxes.xyxy:  # Extract bounding boxes
            x1, y1, x2, y2 = map(int, box)
            current_boxes.append((x1, y1, x2, y2))

    # Update persistent_boxes with the current detections
    new_persistent_boxes = {}
    for box in current_boxes:
        new_persistent_boxes[box] = timeout_frames  # Reset timeout for detected boxes

    # Decrease timeout for old boxes
    for box in persistent_boxes:
        if box not in new_persistent_boxes:  # If not detected now, decrease timeout
            if persistent_boxes[box] > 0:
                new_persistent_boxes[box] = persistent_boxes[box] - 1

    persistent_boxes = new_persistent_boxes  # Update persistent storage

    # Draw all bounding boxes (both new and persistent)
    for box, remaining_time in persistent_boxes.items():
        x1, y1, x2, y2 = box
        color = (0, 255, 0)  # Green color for bounding box
        thickness = 2 if remaining_time > 0 else 1  # Reduce thickness if fading
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, thickness)

    # Write the frame to output video
    out.write(frame)

    # Display the frame (optional)
    cv2.imshow("YOLO Video", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Load the YOLO model
model = YOLO("best.pt")

# Open the video file
video_path = "downloads/fgt.mp4"
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create a VideoWriter object
output_path = "output.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Dictionary to store detected objects (bounding boxes + labels) with a timeout
persistent_objects = {}
timeout_frames = 10  # Keep bounding boxes for 10 frames even if they disappear

frame_count = 0

# Process each frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Stop when video ends

    frame_count += 1

    # Run YOLO inference on the frame
    results = model(frame, conf=0.1)

    # Store new detections
    current_objects = {}
    for result in results:
        for box, cls in zip(result.boxes.xyxy, result.boxes.cls):
            x1, y1, x2, y2 = map(int, box)  # Bounding box coordinates
            label = model.names[int(cls)]  # Get class name
            current_objects[(x1, y1, x2, y2)] = (label, timeout_frames)  # Store with timeout

    # Update persistent_objects with the current detections
    new_persistent_objects = {}
    for box, (label, _) in current_objects.items():
        new_persistent_objects[box] = (label, timeout_frames)  # Reset timeout

    # Decrease timeout for old boxes
    for box, (label, remaining_time) in persistent_objects.items():
        if box not in new_persistent_objects:  # If not detected now, decrease timeout
            if remaining_time > 0:
                new_persistent_objects[box] = (label, remaining_time - 1)

    persistent_objects = new_persistent_objects  # Update persistent storage

    # Draw all bounding boxes (both new and persistent)
    for box, (label, remaining_time) in persistent_objects.items():
        x1, y1, x2, y2 = box
        color = (0, 255, 0)  # Green color for bounding box
        thickness = 2 if remaining_time > 0 else 1  # Reduce thickness if fading
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, thickness)

        # Display label text above the bounding box
        text = f"{label}"  # You can add confidence if needed
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 0.5
        text_thickness = 1
        text_size, _ = cv2.getTextSize(text, font, font_scale, text_thickness)
        text_x, text_y = x1, y1 - 5

        # Create a filled rectangle for better text visibility
        cv2.rectangle(frame, (text_x, text_y - text_size[1] - 2), (text_x + text_size[0], text_y + 2), (0, 255, 0), -1)
        cv2.putText(frame, text, (text_x, text_y), font, font_scale, (0, 0, 0), text_thickness)

    # Write the frame to output video
    out.write(frame)

    # Display the frame (optional)
    cv2.imshow("YOLO Video", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
```
